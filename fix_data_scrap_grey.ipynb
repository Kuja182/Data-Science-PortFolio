```python
data_lc = pd.read_csv("./data/csv_scrap/grey/lentillas_grey.csv", index_col=0) # En este csv tendremos que sacar los liquidos ya que están dentro del mismo
data_gafas = pd.read_csv("./data/csv_scrap/grey/toda-la-coleccion-graduadas_greyhounders.csv", index_col=0)
```

```python
data_lc["Empresa"] = "Greyhounders"
data_lc["Categoria"] = "Lentes de contacto"
data_lc["Producto"] = "Biofinty"
data_lc["Reemplazo"] = "Mensual"

data_lc = data_lc.rename(columns={"nombre_completo":"Nombre completo",
                                  "precio":"Precio"})
data_lc = data_lc[["Empresa","Nombre completo","Categoria","Producto","Reemplazo","Precio"]]
```

```python
data_lc.loc[1][4] = "Diarias"
data_lc.loc[4][4] = "Diarias"
data_lc.loc[1][3] = "MyDay"
data_lc.loc[4][3] = "MyDay"
data_lc
```

```python
for indice, valor in data_lc["Precio"].iteritems(): # Con este codigo limpiamos la fila de precio de cualquier tipo de carcarter, convirtiendo la cadena en float.
    cadena = valor
    for caracter in cadena:
        try:
            valor_float = float(cadena)
            data_lc.at[indice, "Precio"] = valor_float
            break
        except ValueError:
            if not caracter.isdigit() and caracter != '.' and caracter != '-':
                cadena = cadena.replace(caracter, '')
            cadena = cadena.replace(',', '.')
data_lc["Precio"] = data_lc["Precio"].astype(float)
```
